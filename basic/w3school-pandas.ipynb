```python
import pandas as pd
import io
import requests

url="https://www.w3schools.com/python/data.csv"

s = requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))

print(df.to_string())
```

```python
print(df.duplicated())
```

```python
df.drop_duplicates(inplace = True)
```

```python
print(df.to_string)
```

```python
print(df.corr())
```

```python
import matplotlib.pyplot as plt
df.plot()
plt.show()
```

```python
sum_column = df.sum(axis=0)
print(sum_column)
```

```python
count_column = df.count(axis=0)
print(count_column)
```
